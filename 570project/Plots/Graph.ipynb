```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
fg_size = {"w":8.0, "h":6.0, "dpi":150}
markersize = 8
```

```python
def size_time(method_name):
    dBasic = np.genfromtxt("output_b.txt", delimiter=',')
    sizeB = dBasic[:,0]
    timeB = dBasic[:,1]
    dDC = np.genfromtxt("output_e.txt", delimiter=',')
    sizeD = dDC[:,0]
    timeD = dDC[:,1]
    
    np.set_printoptions(precision=5, suppress=True)
    
    print(sizeB)
    print(timeB)
    print(sizeD)
    print(timeD)
    
    plt.clf()
    plt.figure(figsize=(fg_size["w"], fg_size["h"]), dpi = fg_size["dpi"])
    b_label = "Basic version"
    d_label = "Memory-efficient"
#     plt.semilogx(sizeB,timeB,label=b_label, base=2)
#     plt.semilogx(sizeD,timeD,label=d_label, base=2)
    plt.plot(sizeB,timeB,label=b_label)
    plt.plot(sizeD,timeD,label=d_label)
    plt.legend()
    plt.xlabel("Problem Size(m+n)")
    plt.ylabel("CPU Time")
    plt_title = "CPU Time VS Problem Size"
    plt.title(plt_title)
    plt.show()
```

```python
def mem_time(method_name):
    dBasic = np.genfromtxt("output_b.txt", delimiter=',')
    sizeB = dBasic[:,0]
    memB = dBasic[:,2]
    dDC = np.genfromtxt("output_e.txt", delimiter=',')
    sizeD = dDC[:,0]
    memD = dDC[:,2]
    
    np.set_printoptions(precision=5, suppress=True)
    
    print(sizeB)
    print(memB)
    print(sizeD)
    print(memD)
    
    plt.clf()
    plt.figure(figsize=(fg_size["w"], fg_size["h"]), dpi = fg_size["dpi"])
    b_label = "Basic version"
    d_label = "Memory-efficient"
#     plt.semilogx(sizeB, memB, label=b_label, base=2)
#     plt.semilogx(sizeD, memD, label=d_label, base=2)
    plt.plot(sizeB, memB, label=b_label)
    plt.plot(sizeD, memD, label=d_label)
    plt.legend()
    plt.xlabel("Problem Size(m+n)")
    plt.ylabel("Memory Usage")
    plt_title = "Memory Usage VS Problem Size"
    plt.title(plt_title)
    plt.show()
```

```python
size_time("Memory usage vs problem size")
mem_time("Memory usage vs problem size")
```
